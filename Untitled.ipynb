```python
ridge_train_pred = []
ridge_test_pred = []
rmes_ridge = 20
alpha_ridge = 0
collect = 20
for alpha in np.arange(0, 100, 0.1):
    x_ridge=dp[['LT-001','DFT-001','PT-004']][3000:7000]
    y_ridge=dp[['FT-002']][3000:7000]
    clf = Ridge(alpha=alpha)
    clf.fit(x_ridge, y_ridge)
    Xtest_ridge=dp[['LT-001','DFT-001','PT-004']][:7200]
    Ytest_ridge=df[['FT-002']][:7200]
    # training
    Ytest_ridge['predict'] = clf.predict(Xtest_ridge)
    y_ridge['train'] = clf.predict(x_ridge)
    test= clf.predict(dp[['LT-001','DFT-001','PT-004']][:3000])
    # prediction
    mes = mean_squared_error(y_ridge['FT-002'],y_ridge['train'])
    rmes = math.sqrt(mes)
    ridge_train_pred.append(rmes)
    if rmes<collect:
        rmes_ridge = rmes
        alpha_ridge = alpha
    mes = mean_squared_error(dp['FT-002'][:3000],test)
    rmes = math.sqrt(mes)
    ridge_test_pred.append(rmes)
```

```python
plt.figure(figsize=(15,3))
plt.plot(ridge_train_pred,'b')
plt.show()
```
